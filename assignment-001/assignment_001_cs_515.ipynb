```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Set seed for reproducibility
np.random.seed(42)

# Generate data for Class 1
x1_class1 = np.random.normal(6, 2.5, 500)
x2_class1 = np.random.normal(12, 2.5, 500)
class1_labels = np.zeros(500)

# Generate data for Class 2
x1_class2 = np.random.normal(14, 2.5, 500)
x2_class2 = np.random.normal(8, 2.5, 500)
class2_labels = np.ones(500)

# Combine the data into one dataset
x1 = np.concatenate([x1_class1, x1_class2])
x2 = np.concatenate([x2_class1, x2_class2])
labels = np.concatenate([class1_labels, class2_labels])

# Create a DataFrame for better structure
data = pd.DataFrame({
    'x1': x1,
    'x2': x2,
    'label': labels
})

# Plot the data
plt.figure(figsize=(8,6))
plt.scatter(data[data['label'] == 0]['x1'], data[data['label'] == 0]['x2'], color='blue', label='Class 1', alpha=0.5)
plt.scatter(data[data['label'] == 1]['x1'], data[data['label'] == 1]['x2'], color='red', label='Class 2', alpha=0.5)
plt.title('Synthetic Two-Class Dataset')
plt.xlabel('Feature x1')
plt.ylabel('Feature x2')
plt.legend()
plt.grid(True)
plt.show()
```
